```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import cryptocompare
import requests
from datetime import datetime 
import json

with open('data.json') as f:
  data = json.load(f)

%matplotlib notebook

plt.style.use('seaborn')

x_vals = []
y_vals = []

def data_importing(data):
    response = requests.get("https://blockchain.info/ticker")
    data = response.json()

def animate(i):
    timestamp = datetime.now()
    timestamp = timestamp.strftime("%H:%M:%S")
    x_vals.append(timestamp)
    y_vals.append(data['USD']["15m"])

    plt.cla()
    
    plt.title('Bitcoin (BTC) Price Live Plotting')
    plt.gcf().canvas.set_window_title('Live Plotting Cryptocurrency')
    
    plt.xlabel('Time')
    plt.xticks(rotation=45)
    plt.ylabel('Price($)')
    plt.plot_date(x_vals,y_vals,linestyle="solid",ms=0)
    plt.tight_layout()
    
   
ani = FuncAnimation(plt.gcf(), animate, interval=10000)

plt.show()
```
